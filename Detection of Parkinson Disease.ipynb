```python
#import library
import pandas as pd
import numpy as np
```

```python
#load data
df = pd.read_csv("parkinsons.data")
df
```

```python
df.head(23)
```

```python
feature = df.loc[:,df.columns != 'status'].values[:,1:]
feature
```

```python
labels = df.loc[:,'status'].values
labels
```

```python
labels[labels==1].shape[0]
```

```python
labels[labels==0].shape[0]
```

```python
#Why only use minMaxScaler in feature beacuse the features have different range of values
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1,1))
x = scaler.fit_transform(feature)
y = labels
```

```python
x
```

```python
y
```

```python
#split it into train ans test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
#train the model
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
y_pred = xgb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#percentage
round(accuracy_score(y_test,y_pred),4)*100
```
